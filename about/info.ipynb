## 1. Level Design

### 1.1 Notation

$\mathcal{X}_{i,j}$ refers to the domain in which the operator $T_{i,j}$ acts on and $\mathcal{Y}_{i,j}$ is the co-domain, for instance if for some $1,j$, $\mathcal{X}_{1,j}=\{1,2\}$ then the operator $$T_{i,j}: \mathcal{X}_{i,j}\times \mathcal{X}_{i,j} \to \mathcal{Y}_{i,j}$$ suppose that specific operator is a plus then $\mathcal{Y}_{1,j} = \{1,2,3,4\}$. 

Note the index $j$ refers to the underlying domain and co-domain for the type of operator $i$.

### 1.2 Speed

 
Difficulty `d` ranges 5–10. Crossing time in seconds:
 
$$speed(d) = 21.5 - 1.5d$$
 
Pixel velocity scales with canvas height `H`:
 
$$v = \frac{H}{speed(d)} \quad \text{px/sec}$$
 
Spawn interval: $\quad interval = speed(d) \times 400 \text{ ms}$
 
 
Crossing **time** is resolution-independent — only px/sec changes. Scale these with `H`:
 
| Element | Formula |
|---|---|
| Slime radius | $r = H / 14.4$ |
| Font size | $fs = H \times 0.027$ |
| Max slimes | $\lfloor H / 200 \rfloor$ |
| Spawn interval | $speed(d) \times 400 \times \frac{520}{H}$ |


### 1.3 Levels

Note that the levels stack in the sense that the set of possible operations and the cardinality associated with the operators increases.

For instance in Level 1-2 is carried over to Level 3-5, however those are only fillers, meaning that only occationally the questions that stem from simpler sets appear later on.

The current scheme is hence solely based on the above, for there is nothing in previous sets that are not included in the current state.
